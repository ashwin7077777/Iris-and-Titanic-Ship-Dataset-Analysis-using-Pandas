```python
ts=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
ts
```

```python
ts.isnull().sum()
```

```python
ts.describe()
```

```python
ts.head()
```

```python
ts.tail()
```

```python
#filling age values with mean value
ts['Age'].fillna(ts['Age'].mean(),inplace=True)
```

```python
ts.info()
```

```python
#Cabin removed
ts.drop('Cabin',axis=1,inplace=True)
```

```python
ts.info()
```

```python
#survival Percentage:
ts['Survived'].mean()*100
```

```python
#survival rate by gender
ts.groupby('Sex')['Survived'].mean()*100
```

```python
#survival rate by Pclass
ts.groupby('Pclass')['Survived'].mean()*100
```

```python
#survival rate by Age
ts.groupby('Survived')['Age'].mean()
```

```python
# People with age lesser than 16 Survival percentage
ts[ts['Age']<16]['Survived'].mean() *100
```

```python
#Survived people arranged in descending order of Age

ts[ts['Survived']==1].sort_values(by='Age',ascending=False)
```

```python
#Category the age into 0 16 60 and more
bins=[0,16,60,float('inf')]
labels=['Child','Adult','Senior']

ts['AgeGroup']=pd.cut(ts['Age'],bins=bins,labels=labels,right=False)
```

```python
ts
```

```python
ts.groupby('AgeGroup')['Survived'].mean()*100
```
